```python
from bs4 import BeautifulSoup
import urllib.request
import os
```

```python
URL = "https://www.dr.com.tr/kitap/spinozanin-sevinci-nereden-geliyor/cetin-balanuye/felsefe/felsefe-bilimi/urunno=0001691317001"
```

```python
page = urllib.request.urlopen(URL)
print(page)
```

```python
soup = BeautifulSoup(page, "html.parser")
print(soup)
```

```python
for link in soup.findAll('a'):
  theLink = link.get('href')
  name = link.string

  print(theLink)
  print(name)
```
